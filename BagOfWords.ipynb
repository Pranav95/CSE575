```python
stopwords = list()
#load all the stop words in a list
with open('stopwords.txt', 'r') as f:
    for line in f:
        stopwords.append(line.strip())
```

```python
import os
bag_of_words = list()
os.chdir("/home/pranav/Statistical Machine Learning/movie_review_data/neg")
path = "/home/pranav/Statistical Machine Learning/movie_review_data/neg"
for filename in os.listdir(path):
    with open(filename, 'r') as f:
        for line in f:
            line = line.split()
            for word in line:
                if word.lower() in bag_of_words:
                    continue
                if word.lower() in stopwords:
                    continue
                else:
                    bag_of_words.append(word.lower())
            
```
